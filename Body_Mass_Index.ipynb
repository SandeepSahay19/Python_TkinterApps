```python
import tkinter as tk

root = tk.Tk()
root.title("BMI Calculator")
root.geometry("350x320")
root.configure(bg="black")

# -------- FUNCTION --------

def calculate_bmi():
    try:
        weight = float(weight_entry.get())
        height = float(height_entry.get())

        bmi = weight / (height * height)

        if bmi < 18.5:
            category = "Underweight"
            color = "cyan"
        elif bmi < 25:
            category = "Normal"
            color = "green"
        elif bmi < 30:
            category = "Overweight"
            color = "orange"
        else:
            category = "Obese"
            color = "red"

        result_label.config(
            text=f"BMI: {bmi:.2f}",
            fg=color
        )

        category_label.config(
            text=f"Category: {category}",
            fg=color
        )

    except:
        result_label.config(text="Invalid Input", fg="red")
        category_label.config(text="")

# -------- TITLE --------

title = tk.Label(
    root,
    text="BMI Calculator",
    font=("Arial",20,"bold"),
    bg="black",
    fg="white"
)
title.pack(pady=10)

# -------- WEIGHT --------

tk.Label(root,text="Weight (kg)",bg="black",fg="white").pack()
weight_entry = tk.Entry(root,font=("Arial",12))
weight_entry.pack(pady=5)

# -------- HEIGHT --------

tk.Label(root,text="Height (meters)",bg="black",fg="white").pack()
height_entry = tk.Entry(root,font=("Arial",12))
height_entry.pack(pady=5)

# -------- BUTTON --------

button = tk.Button(
    root,
    text="Calculate BMI",
    command=calculate_bmi,
    bg="#4CAF50",
    fg="white",
    font=("Arial",12),
    width=15
)
button.pack(pady=15)

# -------- RESULT --------

result_label = tk.Label(
    root,
    text="",
    font=("Arial",16,"bold"),
    bg="black"
)
result_label.pack()

category_label = tk.Label(
    root,
    text="",
    font=("Arial",14),
    bg="black"
)
category_label.pack()

root.mainloop()
```
